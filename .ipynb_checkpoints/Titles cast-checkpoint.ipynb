```python
import numpy as np
import pandas as pd
```

```python
titles = pd.read_csv('titles.csv')
```

```python
titles
```

```python
titles = titles[titles['type']=='Movie']
```

```python
titles = titles.reset_index()
titles
```

```python
def cast_split(names):
    if ',' in names:
        return names.split(",")

titles['people'] = titles['cast'].apply(cast_split)
titles
```

```python
titles.nunique()
```

```python
titles['cast'].value_counts().head(10)
```

```python
titles2 = pd.read_csv('titles.csv')
```

```python
titles2
```

```python
titles2 = titles2.drop(['show_id','title','director','rating', 'duration','description'],axis = 1)
titles2
```

```python
titles2 = titles2.dropna(axis = 0, thresh = 4)
titles2 = titles2[titles2['listed_in']!= 'Documentaries']
titles2 = titles2[titles2['listed_in']!= 'Stand-Up Comedy']
titles2 = titles2[titles2['listed_in']!= 'Dramas, International Movies']
titles2 = titles2[titles2['listed_in']!= 'Dramas, Independent Movies, International Movies']
titles2 = titles2[titles2['listed_in']!= 'Comedies, Dramas, International Movies']
titles2 = titles2[titles2['listed_in']!= 'Kids\' TV']
titles2 = titles2[titles2['listed_in']!= 'Documentaries, International Movies']
titles2 = titles2[titles2['release_year']>= 2010]
titles2 = titles2[titles2['type']!= 'TV Show']
```

```python
titles2 = titles2[titles2['country']=='United States']
titles2
```
