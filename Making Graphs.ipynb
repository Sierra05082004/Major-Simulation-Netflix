```python
rating_scores = pd.read_excel('Movie rating score.xlsx')
rating_scores = rating_scores.drop('Unnamed: 0', axis = 1)
rating_scores.sort_values('user rating score', ascending = True)
rating_scores = rating_scores[rating_scores['user rating score']>=95]
rating_scores = rating_scores.reset_index()
rating_scores
```

```python
rating_average = pd.read_excel('Movie rating average.xlsx')
rating_average =rating_average.drop('Unnamed: 0', axis = 1)
rating_average
```

```python
rating_average = rating_average.sort_values(['vote_average', 'vote_count'], ascending = False)
rating_average = rating_average.reset_index()
rating_average
```

```python
rating_average = rating_average[rating_average['vote_average']>= 8.0]
rating_average = rating_average.drop('index', axis = 1)
```

```python
rating_average = rating_average[rating_average['vote_average']>= 8.2]
rating_average
```
