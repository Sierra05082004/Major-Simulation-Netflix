```python
import numpy as np
import pandas as pd
```

```python
ratings = pd.read_excel('Movie ratings.xlsx')
ratings = ratings.drop(['Unnamed: 0','user rating score', 'user rating size', 'count'], axis = 1)

ratings['cast'].value_counts().head(17)
counts = ratings['cast'].value_counts()
ratings = ratings[ratings['cast'].isin(counts[counts>1].index) ]
ratings = ratings.reset_index(drop = True)
ratings['title'].value_counts().to_excel('Most repeated movie actors.xlsx')
```
